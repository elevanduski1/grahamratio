```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
def getData(ticker):
    r=requests.get("https://finviz.com/quote.ashx?t={}".format(ticker)).content
    data=bs(r,"html.parser")
    
    # book/sh
    tempList = data.find_all("td")
    text1 = str(tempList[92])
    text2 = text1.split("<b>")[1]
    text3 = text2.split("</b>")[0]
    bvps = float (text3)
    
    #eps
    tempList = data.find_all("td")
    text1 = str(tempList[48])
    text2 = text1.split("<b>")[1]
    text3 = text2.split("</b>")[0]
    eps = float (text3)
    
    #currentPrice
    tempList = data.find_all("td")
    text1 = str(tempList[174])
    text2 = text1.split("<b>")[1]
    text3 = text2.split("</b>")[0]
    price = float (text3)
 
    #divYield
    tempList = data.find_all("td")
    text1 = str(tempList[116])
    text2 = text1.split("<b>")[1]
    text3 = text2.split("</b>")[0]
    dividend = float (text3)
    
    #graham number and ratio
    grahamNumber = ((22.5*bvps*eps)**(1/2))
    grahamRatio = grahamNumber/price
    
    return grahamRatio
```

```python
getData('FITB')
```
